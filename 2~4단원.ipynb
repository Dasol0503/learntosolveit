##4.3.1이항분포에 관한 함수

```python
from scipy.stats import binom #scipy.stats 패키지(binom) 불러오기
```

```python
binom.pmf(3,10,0.5) #동전을 10번 던져서 앞면이 6번 이하로 나올 확률
```

```python
binom.cdf(6,10,0.5) #동전을 10변 던져서 앞면이 6번 이하로 나올 확률
```

```python
1-binom.cdf(6,10,0.5) #P(X<6) = 1-P
```

```python
binom.ppf(0.3,10,0.5) #누적확률이 0.3이상이 되는 x
```

```python
binom.rvs(10,0.5,size=30) #동전을 10번 던지는 실험을 30번 반복
```

```python
hypergeom.pmf(10,1000,200,50)  #N=1, A=200, n=50 -> P(X=10)
```

```python
hypergeom.cdf(10,1000,200,50)  #N=1, A=200, n=50 -> P(X=10)
```

```python
1-hypergeom.cdf(10,1000,200,50) #P(x<10) = 1-P(x<=10)
```

```python
hypergeom.ppf(0.5,1000,200,50) #누적확률이 0.5 이상이 되는 x
```

```python
hypergeom.rvs(1000,200,50,size=25) #모의실험을 25번 반복하는 경우
```

```python
from scipy.stats import binom # scipy.stats 패키지 (binom) 이항분포 함수 불러오기

n = 4
p = 0.5
k = list(range(7))

fig, ax = plt.subplots() # 그래프 합침
plt.plot(k,binom.pmf(k,n,p)) # 확률질량함수(이항분포) 선 그래프 
ax.stem(k,binom.pmf(k,n,p)) # 확률질량함수(이항분포) 꺾은 선 그래프
plt.show()
```

```python
from scipy.stats import hypergeom # scipy.stats 패키지 (hypergeom) 초기하분포 함수 불러오기

N = 100
A = 30
n = 10
k = list(range(9))

fig, ax = plt.subplots() # 그래프 합침
plt.plot(k,hypergeom.pmf(k,N,A,n)) # 초기하분포 선 그래프 
ax.stem(k,hypergeom.pmf(k,N,A,n)) # 초기하분포 꺾은 선 그래프
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm # scipy.stats 패키지 (norm) 정규분포 함수 불러오기

ax1_mu = 10
ax1_sigma = 5

ax2_mu = 20
ax2_sigma = 10

k = np.arange(0,40,0.1) 

fig, ax = plt.subplots() # 그래프 합침
plt.plot(k, norm.pdf (k,ax1_mu, ax1_sigma)) # ax1의 정규분포 선 그래프
plt.plot(k, norm.pdf (k,ax2_mu, ax2_sigma)) # ax2의 정규분포 선 그래프
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import t # scipy.stats 패키지 (t) t-분포 함수 불러오기

ax1_v = 30

ax2_v = 1

k = np.arange(-4.5,4.5,0.1) 

fig, ax = plt.subplots() # 그래프 합침
plt.plot(k, t.pdf (k,ax1_v)) # ax1의 t-분포 선 그래프
plt.plot(k, t.pdf (k,ax2_v), linestyle='--') # ax2의 t-분포 선 그래프
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import chi2 # scipy.stats 패키지 (chi2) 카이제곱분포 함수 불러오기

ax1_v = 3

ax2_v = 6

k = np.arange(0,10,0.1) 

fig, ax = plt.subplots() # 그래프 합침
plt.plot(k, chi2.pdf (k,ax1_v))  # ax1의 카이제곱분포 선 그래프
plt.plot(k, chi2.pdf (k,ax2_v), linestyle='--') # ax2의 카이제곱분포 선 그래프
plt.show()
```
